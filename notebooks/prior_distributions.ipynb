### <center> Exploration of the Influence of Prior Distributions

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta
```

```python
a = 2.
b = 10.
plt.title(f'Beta Distribution Expectation: {a/(a+b)}')
plt.plot(np.linspace(0.,1.,100),beta.pdf(a = a, b = b, x = np.linspace(0.,1.,100)))
plt.show()
    
```
