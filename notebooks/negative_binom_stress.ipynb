```python
import sys
sys.path.append('../')

import numpy as np
import matplotlib.pyplot as plt

from math_utils import nbinom_logpmf
```

```python
means = np.arange(0,50,5)
R = 10

evals = np.arange(0,100,1)

plt.title(f'Negative Binomial PMF with Various Means and Fixed R of {R}')
for i in range(len(means)):
    probs = nbinom_logpmf(evals,n = np.array([R]),p = R/(R + means[i]))
    plt.plot(evals,np.exp(probs),label = f'mean: {means[i]}')
plt.legend()
```

```python
mean = 50
R = np.arange(0,500,50)

evals = np.arange(0,100,1)

plt.title(f'Negative Binomial PMF with Various R and Fixed Mean of {mean}')
for i in range(len(R)):
    probs = nbinom_logpmf(evals,n = np.array([R[i]]),p = R[i]/(R[i] + mean))
    plt.plot(evals,np.exp(probs),label = f'R: {R[i]}')
plt.legend()
```
