```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(1)
dt = 0.00001
t_end = 20
t = np.arange(0,t_end+dt,dt)

state = np.zeros((2,len(t)))
state[:,0] = np.sqrt(2)/2
```

```python
for i_time in range(0, len(t)-1):
    x,y = state[:,i_time]
    dW = rng.normal(0,np.sqrt(dt))
    dx = -y * dW - 1/2 * x * dt
    dy = x * dW - 1/2 * y * dt
    state[0,i_time+1] = x + dx
    state[1,i_time+1] = y + dy
```

```python
plt.plot(state[0,:],state[1,:])
```

```python
plt.plot(t,state[0,:])
```
